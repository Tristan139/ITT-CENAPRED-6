```python
import pandas as pd
import geopandas as gpd

df = pd.read_csv("SSNMX_cat.csv")
df
```

```python
faults = gpd.read_file('gem-global-active-faults/geojson/gem_active_faults.geojson')
faults.to_csv("faults.csv")
faults
```

```python
faults["geometry"].plot()
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
mexico = world[world.name=="Mexico"]
mexico.plot()
```

```python
ax = mexico.plot()
x1,x2,y1,y2 = ax.axis()

df.plot.scatter(ax=ax, x="Longitud", y="Latitud", s=5**df["Magnitud"]/100, c="Profundidad", cmap="jet_r", alpha=0.5, figsize=(10,10))
ax.set_xlim((x1+x2)/2, x2)
ax.set_ylim(y1, (y1+y2)/2)

faults.geometry.plot(ax=ax)
mexico.plot(ax=ax, alpha=0.3)
```

```python
ax.figure.savefig("mexquakes.png")
```
